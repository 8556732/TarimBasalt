# Isopach map histogram (KDE)

#### This notebook generates Figure 4 of the manuscript

This takes a while to run, please be patient. 

#### Notes:
The input file "MergedThicknessData.csv" contains the values of modelled thicknesses without coordinates, this was done to reduce the plotting time for the figure. If you would like to access the data used in this file it can be found as Zmap+ Grid format (.dat) or XYZ .csv files within this folder. 

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Arial"
plt.rcParams.update({'font.size': 14})
plt.rcParams['figure.figsize'] = [12, 8]

df = pd.read_csv("MergedThicknessData.csv");
ax = df.plot.kde();
plt.xlim(0,1000);
```
